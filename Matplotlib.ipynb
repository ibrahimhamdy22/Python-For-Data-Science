### Scatter Plot 

#### Purpose: Visualize the relamtionship between two numerical variables.


```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
x = [1, 2, 3, 4, 5]
y = [10, 12, 5, 8, 15]

plt.scatter(x,y)
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title ('catter Plot Example')
```

### Line Plot  

#### Purpose: show the trend or change in data over a continous interval


```python
x = [1, 2, 3, 4, 5]
y = [10, 12, 5, 8, 15]
plt.plot(x,y)
plt.xlabel('X-label')
plt.ylabel('Y-label')
plt.title('Line plot Example')
```

### subplots 

#### Purpose: create multiple plots in the same figure.


```python
x = [1, 2, 3, 4, 5]
y1 = [10, 12, 5, 8, 15]
y2 = [5, 8, 6, 12, 9]

plt.subplot(1,2,1)
plt.plot(x,y1)
# plt.title('subplot 1')

plt.subplot(1,2,2)
plt.plot(x,y2)
# plt.title('subplt2')
```

### Bar Plot 

#### Purpose: Display categrorical data using rectangular bars


```python
categories = ['Category A', 'Category B', 'Category C']
values = [10, 15, 7]

plt.bar(categories,values)
plt.xlabel('categories')
plt.ylabel('Values')
```

### Histogram 

#### Purpose: visualize the distribution of a single numerical variable


```python
data = [1, 2, 2, 3, 3, 3, 4, 4, 5, 5]

plt.hist(data, bins=5)
plt.xlabel('Values')
plt.ylabel('Frequency')
plt.title('Histogram Example')
```

### Box Plot (Box and Whisker plot) 

#### Purpose: show the distribution and statistical summary about the dataset


```python
data = [1, 2, 2, 3, 3, 3, 4, 4, 5, 5]
plt.boxplot(data , vert = True , patch_artist = True)
plt.ylabel('Values')
plt.title('Box Plot Example')
```

### Pie Chart

#### Purpose: Display the composition of a whole as a set of slices


```python
labels = 'python','C++','Ruby','Java'
sizes = [215,130,245,210]
colors = ['gold','yellowgreen','lightcoral','lightskyblue']
explode = (0.4,0,0,0)   # how far is the first slice from the others

# plot 

plt.pie(sizes, explode=explode, labels=labels, colors= colors,
      shadow=True
       )
plt.axis('equal')
```
